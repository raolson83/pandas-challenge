```python
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
school_data_to_load = Path("C:/Users/raols/OneDrive/Documents/GitHub/pandas-challenge/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("C:/Users/raols/OneDrive/Documents/GitHub/pandas-challenge/PyCitySchools/Resources//students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
                           
# Combine the data into a single dataset.  
sch_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
sch_df.head() 
```

```python
# Calculate the average (mean) math score
average_math_score = sch_df["math_score"].mean()
average_math_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = sch_df[(sch_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = sch_df[(sch_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Calculate the total student count per school
per_school_counts = sch_df['school_name'].value_counts()
per_school_counts = pd.DataFrame(per_school_counts)
per_school_counts
```

```python
# Calculate the average test scores per school
per_school_math = sch_df.groupby(['school_name'])['math_score'].mean()
per_school_math= pd.DataFrame(per_school_math)
per_school_math.columns = ['Avg Math Score']

per_school_reading= sch_df.groupby(['school_name'])['reading_score'].mean()
per_school_reading= pd.DataFrame(per_school_reading)
per_school_reading.columns = ['Avg Reading Score']
```

```python
per_school_math
```

```python
# Calculate the number of students per school with math scores of 70 or higher

school_students_passing_math = sch_df.groupby(['school_name'])["math_score"].apply(lambda x: x[x >= 70].count())
school_students_passing_math = pd.DataFrame(school_students_passing_math)
school_students_passing_math.columns = ['Students Passing Math']
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher

school_students_passing_reading = sch_df.groupby(['school_name'])["reading_score"].apply(lambda x: x[x >= 70].count())
school_students_passing_reading = pd.DataFrame(school_students_passing_reading)
school_students_passing_reading.columns = ['Students Passing Reading']
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = sch_df[
    (sch_df["reading_score"] >= 70) & (sch_df["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()

school_students_passing_math_and_reading = pd.DataFrame(school_students_passing_math_and_reading)
school_students_passing_math_and_reading.columns = ['Students Passing Math and Reading']
school_students_passing_math_and_reading

```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math['Students Passing Math'] / per_school_counts['count'] * 100
per_school_passing_math = pd.DataFrame(per_school_passing_math)
per_school_passing_math.columns = ['% Passing Math']

per_school_passing_reading = school_students_passing_reading['Students Passing Reading'] / per_school_counts['count'] * 100
per_school_passing_reading = pd.DataFrame(per_school_passing_reading)
per_school_passing_reading.columns = ['% Passing Reading']

overall_passing_rate = school_students_passing_math_and_reading['Students Passing Math and Reading'] / per_school_counts['count'] * 100
overall_passing_rate = pd.DataFrame(overall_passing_rate)
overall_passing_rate.columns = ['% Passing Overall']

overall_passing_rate
```

```python
## Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.merge(school_data,per_school_reading, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,school_students_passing_reading, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,per_school_passing_reading, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,per_school_math, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,school_students_passing_math, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,per_school_passing_math, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,school_students_passing_math_and_reading, on = 'school_name', how='outer')
per_school_summary = pd.merge(per_school_summary,overall_passing_rate, on = 'school_name', how='outer')

per_school_summary = per_school_summary.rename(columns={'budget': 'Total School Budget', 'per_capita_spending': "Per Student Budget"})


# # Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# # Display the DataFrame
per_school_summary.head()
```

```python
# Use the code provided to separate the data by grade
ninth_graders = sch_df[(sch_df["grade"] == "9th")]
tenth_graders = sch_df[(sch_df["grade"] == "10th")]
eleventh_graders = sch_df[(sch_df["grade"] == "11th")]
twelfth_graders = sch_df[(sch_df["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grader_math_scores = ninth_graders.groupby(['school_name'])['math_score'].mean().reset_index()
ninth_grader_math_scores = ninth_grader_math_scores.rename(columns={'math_score': "9th Grader Math Score"})

tenth_grader_math_scores = tenth_graders.groupby(['school_name'])['math_score'].mean().reset_index()
tenth_grader_math_scores = tenth_grader_math_scores.rename(columns={'math_score': "10th Grader Math Score"})

eleventh_grader_math_scores = eleventh_graders.groupby(['school_name'])['math_score'].mean().reset_index()
eleventh_grader_math_scores = eleventh_grader_math_scores.rename(columns={'math_score': "11th Grader Math Score"})

twelfth_grader_math_scores = twelfth_graders.groupby(['school_name'])['math_score'].mean().reset_index()
twelfth_grader_math_scores = twelfth_grader_math_scores.rename(columns={'math_score': "12th Grader Math Score"})

# # Combine each of the scores above into single DataFrame called `math_scores_by_grade` MERGE
math_scores_by_grade = pd.merge(ninth_grader_math_scores,tenth_grader_math_scores, on = 'school_name', how='outer')
math_scores_by_grade = pd.merge(math_scores_by_grade,eleventh_grader_math_scores, on = 'school_name', how='outer')
math_scores_by_grade = pd.merge(math_scores_by_grade,twelfth_grader_math_scores, on = 'school_name', how='outer')

# # Minor data wrangling
math_scores_by_grade.index.name = None

# # Display the DataFrame
math_scores_by_grade

```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'].str[1:].astype(float), spending_bins, labels = labels, include_lowest = True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Math Score"].mean().reset_index()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Reading Score"].mean().reset_index()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean().reset_index()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean().reset_index()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean().reset_index()
```

```python
overall_passing_spending
```

```python
# Assemble into DataFrame
spending_summary = pd.merge(spending_math_scores,spending_reading_scores, on='Spending Ranges (Per Student)', how = 'outer')
spending_summary = pd.merge(spending_summary,spending_passing_math, on='Spending Ranges (Per Student)', how = 'outer')
spending_summary = pd.merge(spending_summary,spending_passing_reading, on='Spending Ranges (Per Student)', how = 'outer')
spending_summary = pd.merge(spending_summary,overall_passing_spending, on='Spending Ranges (Per Student)', how = 'outer')

# Display results
spending_summary
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(school_spending_df['size'], size_bins, labels = labels, include_lowest = True)
per_school_summary.head()
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Avg Math Score"].mean().reset_index()
size_reading_scores = per_school_summary.groupby(["School Size"])["Avg Reading Score"].mean().reset_index()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean().reset_index()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean().reset_index()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Passing Overall"].mean().reset_index()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.merge(size_math_scores,size_reading_scores, on='School Size', how = 'outer')
size_summary = pd.merge(size_summary,size_passing_math, on='School Size', how = 'outer')
size_summary = pd.merge(size_summary,size_passing_reading, on='School Size', how = 'outer')
size_summary = pd.merge(size_summary,size_overall_passing, on='School Size', how = 'outer')

# Display results
size_summary

```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["type"])["Avg Math Score"].mean().reset_index()
average_reading_score_by_type = per_school_summary.groupby(["type"])["Avg Reading Score"].mean().reset_index()
average_percent_passing_math_by_type = per_school_summary.groupby(["type"])["% Passing Math"].mean().reset_index()
average_percent_passing_reading_by_type = per_school_summary.groupby(["type"])["% Passing Reading"].mean().reset_index()
average_percent_overall_passing_by_type = per_school_summary.groupby(["type"])["% Passing Overall"].mean().reset_index()
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`
type_summary = pd.merge(average_math_score_by_type,average_reading_score_by_type, on='type', how = 'outer')
type_summary = pd.merge(type_summary,average_percent_passing_math_by_type, on='type', how = 'outer')
type_summary = pd.merge(type_summary,average_percent_passing_reading_by_type, on='type', how = 'outer')
type_summary = pd.merge(type_summary,average_percent_overall_passing_by_type, on='type', how = 'outer')

# Display results
type_summary
```
